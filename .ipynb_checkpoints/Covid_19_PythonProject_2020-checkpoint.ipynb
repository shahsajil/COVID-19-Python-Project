```python
# importing libraries needed for this Project
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import ploty.express as px
```

```python
# importing excel dataset into notebook
df = pd.read_excel ('Covid_19.xlsx')
```

```python
# displaying the dataset 
df
```

```python
# data cleaning replacing the NaN with 0 value
df = df.fillna(0)
```

```python
# displaying dataset after data cleaning 
df 
```

```python
# Adding columns (Total Cases and New Cases) and (Total Deaths and New Deaths)
df["Total Cases (Up To Date)"] = df["Total Cases"]+df["New Cases"]
df["Total Deaths (Up To Date)"] = df["Total Deaths"]+df["New Deaths"]
```

```python
# displaying dataset with new columns
df
```

```python
# diplaying the needed columns only 
totalcases_df = df[["Country, Other", "Total Cases (Up To Date)"]]
```

```python
totalcases_df.sort_values("Total Cases (Up To Date)", ascending=False)
```

```python
totalcases_df[['Country, Other', 'Total Cases (Up To Date)']].set_index ('Country, Other')[:25].plot.bar()
```

```python
# diplaying the needed columns only 
totaldeaths_df = df[["Country, Other", "Total Deaths (Up To Date)"]]
```

```python
totaldeaths_df.sort_values("Total Deaths (Up To Date)", ascending=False)
```

```python
totaldeaths_df[['Country, Other', 'Total Deaths (Up To Date)']].set_index ('Country, Other')[:10].plot.bar()
```

```python
sns.barplot(data=totaldeaths_df[:10],x='Country, Other', y='Total Deaths (Up To Date)')
```
